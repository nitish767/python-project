```python
import seaborn as sns 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
#step 1
tips = sns.load_dataset("tips")

#step 2
print(tips.head())
```

```python
#step 3: selecting features for simple Linear Regression
x = tips[['total_bill']] # independent variable
y = tips['tip'] # dependent variable
```

```python
# Step 4: Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Step 5: Fit a simple linear regression model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Step 6: Get model parameters
m = model.coef_[0]  # slope
b = model.intercept_  # intercept
print(f"Regression Equation: Y = {m:.2f}X + {b:.2f}")
```

```python
# Step 7: Make Predictions
y_pred = model.predict(x_test)

```

```python
# Step 8: Evaluate Model Performance
mse = mean_squared_error(y_test, y_pred)  # Calculate Mean Squared Error
r2 = r2_score(y_test, y_pred)  # Calculate R-squared value

# Print the performance metrics
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```

```python
# Step 9: Visualization of Regression Line
plt.figure(figsize=(8, 5))
plt.scatter(x_test, y_test, color='blue', label="Actual Tips")  # Scatter plot of actual data
plt.plot(x_test, y_pred, color='red', linewidth=2, label="Regression Line")  # Regression line
plt.title("Regression Line: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd

# Load dataset
tips = sns.load_dataset("tips")

# Convert 'sex' to numeric values (0: Female, 1: Male) for visualization
tips['sex_numeric'] = tips['sex'].apply(lambda x: 1 if x == 'Male' else 0)

# Create 3D figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with meaningful axes
ax.scatter(tips['total_bill'], tips['tip'], tips['sex_numeric'], c=tips['sex_numeric'], cmap='coolwarm', alpha=0.7)

# Set axis labels
ax.set_xlabel("Total Bill ($)")  
ax.set_ylabel("Tip ($)")         
ax.set_zlabel("Sex (0 = Female, 1 = Male)")  

plt.title("3D Multiple Regression Visualization")
plt.show()
```
